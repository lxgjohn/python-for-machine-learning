```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r'iris.data.csv', header=None)
#删除不需要的数据列。
data.drop([4], axis=1, inplace=True)
#删除重复的数据。
data.drop_duplicates(inplace=True)
```

```python
class KNN:
    '''用python实现k近邻算法。（回归预测）
    
    该算法用于回归预测，利用前三个特征值，寻找k个邻居，然后再根据k个邻居的特征值，
    预测第四个特征值。
    '''
    def __init__(self, k):
        '''初始化方法。
        
        Parameters
        -----
        k：int
            邻居的个数。
            
        '''
        self.k = k
        
    def fit(self, X, y):
        '''训练的方法。
        
        Parameters
        -----
        X：类数组类型（特征矩阵）。形状为：[样本数量，特征数量]
            待训练的样本特征。（属性）
            
        y：类数组类型（目标标签）。形状为：[样本数量]
            每个样本的目标值。（标签）
        '''
        #注意：将X与y转换成ndarray类型，以便后续统一处理。
        self.X = np.asarray(X)
        self.y = np.asarray(y)
        
    def predict(self, X):
        '''预测的方法。
        
        Parameters
        -----
        X：类数组类型。形状为：[样本数量，特征数量]
            待测试的样本特征。（属性）
            
        Returns
        -----
        result：数组类型。
            预测的结果值。
        '''
        #转换成数组类型。
        X = np.asarray(X)
        #保存预测的结果值。
        result = []
        
        for x in X:
            #计算与训练集每个X的距离。
            dis = np.sqrt(np.sum((x -self.X) ** 2, axis=1))
            #返回数组排序后，每个元素在原数组（排序前的数组）中的索引
            index = dis.argsort()
            #取前k个最近的索引。（原数组中的索引）
            index = index[:self.k]
            #计算均值，然后加入到结果列表中。np.mean 为求均值的运算。
            result.append(np.mean(self.y[index]))  
        return np.asarray(result)
    
    def predict2(self, X):
        '''预测的方法。(考虑权重)
        
        权重的计算方式：每个节点（邻居）距离的倒数/所有节点距离的倒数之和
        
        Parameters
        -----
        X：类数组类型。形状为：[样本数量，特征数量]
            待测试的样本特征。（属性）
            
        Returns
        -----
        result：数组类型。
            预测的结果值。
        '''
        #转换成数组类型。
        X = np.asarray(X)
        #保存预测的结果值。
        result = []
        
        for x in X:
            #计算与训练集每个X的距离。
            dis = np.sqrt(np.sum((x -self.X) ** 2, axis=1))
            #返回数组排序后，每个元素在原数组（排序前的数组）中的索引
            index = dis.argsort()
            #取前k个最近的索引。（原数组中的索引）
            index = index[:self.k]
            #计算所有邻居节点距离的倒数之和。[注意：最后加上一个很小的值，避免除数（距离）为0的情况]
            s = np.sum(1 / (dis[index] + 0.001))
            #计算每个节点距离的倒数，除以所有节点距离倒数之和，得到权重值。
            weight = (1 / (dis[index] + 0.001)) / s
            #使用邻居节点的标签值，乘以权重，，然后相加，得到最终的预测值。
            result.append(np.sum(self.y[index] * weight))
        return np.asarray(result)
```

```python
#创建训练集和测试集
t = data.sample(len(data), random_state=0)
train_X = t.iloc[:120, :-1]
train_y = t.iloc[:120, -1]
test_X = t.iloc[120:, :-1]
test_y = t.iloc[120:, -1]
#调用KNN回归的类。
knn = KNN(k=3)
knn.fit(train_X, train_y)
result = knn.predict(test_X)
#展示数据
display(result)
display(np.mean((result - test_y) ** 2))
display(test_y.values)
```

```python
#调用包含权重的预测方法
result = knn.predict2(test_X)
display(np.mean(result - test_y) ** 2)
```
